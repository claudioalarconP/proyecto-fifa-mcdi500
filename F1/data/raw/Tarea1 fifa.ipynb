# Análisis Exploratorio

Se generan nuevas variables y se exploran relaciones entre los datos para identificar patrones relevantes del torneo.

```python
df['total_goals'] = df['number of goals team1'] + df['number of goals team2']

df[['team1','team2','total_goals']].sort_values(
    by='total_goals',
    ascending=False
).head()
```

```python
df.groupby('team1')['number of goals team1'].sum().sort_values(ascending=False).head(10)
```

# Visualización de Datos

Se construye un histograma para observar la distribución de goles anotados por los equipos durante los encuentros.

```python
import matplotlib.pyplot as plt

df['number of goals team1'].hist()

plt.title('Distribución de goles Team1')
plt.xlabel('Goles')
plt.ylabel('Frecuencia')
plt.show()
```
